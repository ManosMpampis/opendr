```python
import argparse
import torch
from opendr.perception.object_detection_2d import NanodetLearner
from opendr.engine.data import Image
from opendr.perception.object_detection_2d import draw_bounding_boxes
```

```python
nanodet = NanodetLearner(model_to_use="plus_m_320", device="cuda")
```

```python
model = nanodet.model
```

```python
model = model.eval()
```

```python
for name,param in model.named_parameters():
    if param.requires_grad:
        print(name)
```

```python
all_count = sum(param.numel() for name, param in model.named_parameters())
trainable_count = sum(param.numel() for name, param in model.named_parameters() if param.requires_grad == True)
freeze_count = sum(param.numel() for name, param in model.named_parameters() if param.requires_grad == False)
print(f"all parameters are: {all_count}")
print(f"trainable parameters are: {trainable_count}")
print(f"freeze parameters are: {freeze_count}")
```
